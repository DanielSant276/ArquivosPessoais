```python
########### Testando a instalação
# importando a biblioteca de "arrays" numéricos
import numpy as np
# importando a biblioteca fuzzy
import skfuzzy as fuzzy
# importando a biblioteca para os gráficos
import matplotlib.pyplot as plt

## Definição do Universo dos Conjuntos
```

```python
# Universos dos conjuntos
print('Temperatura:')
temperatura = np.arange(0, 29.5, 1)
print(temperatura)
print('Luz:')
luz = np.arange(0, 100.5, 1)
print(luz)
print('Quantidade estimada de turistas:')
turistas = np.arange(0, 100.5, 1)
print(turistas)
```

```python
# Conjunto da Temperatura - Fria, Morna, Quente
TF = fuzzy.trapmf(temperatura, [0, 0, 17, 20])
TM = fuzzy.trapmf(temperatura, [17, 20, 26, 29])
TQ = fuzzy.trapmf(temperatura, [26, 29, 29, 29])
```

```python
# Conjunto da Luz - Nublado, Parcialmente Ensolarado, Ensolarado
LN = fuzzy.trapmf(luz, [0, 0, 30, 50])
LPE = fuzzy.trimf(luz, [30, 50, 100])
LE = fuzzy.trapmf(luz, [50, 100, 100, 100])
```

```python
# Conjunto da Quantidade estimada de turistas - Baixo, Médio, Alto
LB = fuzzy.trimf(turistas, [0, 0, 50])
LM = fuzzy.trimf(turistas, [0, 50, 100])
LA = fuzzy.trimf(turistas, [50, 100, 100])
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))
#
ax0.plot(temperatura, TF, 'b', linewidth=1.5, label='Fria')
ax0.plot(temperatura, TM, 'g', linewidth=1.5, label='Morna')
ax0.plot(temperatura, TQ, 'r', linewidth=1.5, label='Quente')
ax0.set_title('Temperatura')
ax0.legend();
#
ax1.plot(luz, LN, 'b', linewidth=1.5, label='Nublado')
ax1.plot(luz, LPE, 'g', linewidth=1.5, label='Parcialmene ensolarado')
ax1.plot(luz, LE, 'r', linewidth=1.5, label='Ensolarado')
ax1.set_title('Luz')
ax1.legend();
#
ax2.plot(turistas, LB, 'b', linewidth=1.5, label='Baixo')
ax2.plot(turistas, LM, 'g', linewidth=1.5, label='Médio')
ax2.plot(turistas, LA, 'r', linewidth=1.5, label='Alto')
ax2.set_title('Quantidade estimada de turistas')
ax2.legend();
#
# Turn off top/right axes
#for ax in (ax0, ax1, ax2):
#    ax.spines['top'].set_visible(False)
#    ax.spines['right'].set_visible(False)
#    ax.get_xaxis().tick_bottom()
#    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
temperaturaFria = [1, 0.57, 0, 0]
temperaturaMorna = [0, 0.33, 1, 0.7]
temperaturaQuente = [0, 0, 0, 0.22]

tempGraus = [temperaturaFria, temperaturaMorna, temperaturaQuente]

luzNublado = [1, 0.7, 0, 0]
luzParcialEnsola = [0, 0.3, 0.8, 0.3]
luzEnsolar = [0, 0, 0.2, 0.7]

luzGraus = [luzNublado, luzParcialEnsola, luzEnsolar]
```

```python
# Operações
# União
zadehUniaoArr = []
def zadehUniao(temperatura, luz):
    return [np.fmin(temperatura[i][j], luz[i][j]) for i in range(3) for j in range(4)]

zadehUniaoArr = zadehUniao(tempGraus, luzGraus)
print("Zadeh União")
print(zadehUniaoArr)
        
produtoUniaoArr = []
def produtoUniao(temperatura, luz):
    return [temperatura[i][j] * luz[i][j] for i in range(3) for j in range(4)]

produtoUniaoArr = produtoUniao(tempGraus, luzGraus)
print("Produto União")
print(produtoUniaoArr)

mediaUniaoArr = []
def mediaUniao(temperatura, luz):
    return [(temperatura[i][j] + luz[i][j]) / 2 for i in range(3) for j in range(4)]

mediaUniaoArr = mediaUniao(tempGraus, luzGraus)
print("Media União")
print(mediaUniaoArr)

somaDiferencaUniaoArr = []
def somaDiferencaUniao(temperatura, luz):
    return [np.fmax(0, temperatura[i][j] + luz[i][j] - 1) for i in range(3) for j in range(4)]

somaDiferencaUniaoArr = somaDiferencaUniao(tempGraus, luzGraus)
print("Soma e diferença limitadas União")
print(somaDiferencaUniaoArr)

# Interseção
zadehIntersacaoArr = []
def zadehIntersacao(temperatura, luz):
    return [np.fmax(temperatura[i][j], luz[i][j]) for i in range(3) for j in range(4)]

zadehIntersacaoArr = zadehIntersacao(tempGraus, luzGraus)
print("Zadeh Interseção")
print(zadehIntersacaoArr)
        
produtoIntersacaoArr = []
def produtoIntersacao(temperatura, luz):
    return [(temperatura[i][j] + luz[i][j]) - (temperatura[i][j] + luz[i][j]) for i in range(3) for j in range(4)]

produtoIntersacaoArr = produtoIntersacao(tempGraus, luzGraus)
print("Produto Interseção")
print(produtoIntersacaoArr)

mediaIntersacaoArr = []
def mediaIntersecao(temperatura, luz):
    return [(2 * np.fmin(temperatura[i][j], luz[i][j]) + 4 * np.fmax(temperatura[i][j], luz[i][j])) / 6 for i in range(3) for j in range(4)]

mediaIntersacaoArr = mediaIntersecao(tempGraus, luzGraus)
print("Media Interseção")
print(mediaIntersacaoArr)

somaDiferencaIntersacaoArr = []
def somaDiferencaIntersacao(temperatura, luz):
    return [np.fmin(1, temperatura[i][j] + luz[i][j]) for i in range(3) for j in range(4)]

somaDiferencaIntersacaoArr = somaDiferencaIntersacao(tempGraus, luzGraus)
print("Soma e diferença limitadas Interseção")
print(somaDiferencaIntersacaoArr)
```

Regra Fuzzy

```python
# Valores fixo 
temperaturaValor = 19
luzValor = 45
```

```python
# Ativar a regras
# Temperatura
temperatura_tf = fuzzy.interp_membership(temperatura, TF, temperaturaValor)
temperatura_tm = fuzzy.interp_membership(temperatura, TM, temperaturaValor)
temperatura_tq = fuzzy.interp_membership(temperatura, TQ, temperaturaValor)

# Luz
luz_ln = fuzzy.interp_membership(luz, LN, luzValor)
luz_lpe = fuzzy.interp_membership(luz, LPE, luzValor)
luz_le = fuzzy.interp_membership(luz, LE, luzValor)
```

```python
print(temperatura_tf, temperatura_tm, temperatura_tq)
```

```python
print(luz_ln, luz_lpe, luz_le)
```

```python
# Indices de quantidades de turistas
# Baixo = [5, 20, 35, 40]
baixo = [0.9, 0.6, 0.3, 0.2]
# Médio = [30, 40, 60, 70]
medio = [0.6, 0.8, 0.8, 0.6]
# Grande = [60, 70, 80, 90]
grande = [0.2, 0.4, 0.6, 0.8]
```

```python
# Parte um da regra
r1Uniao = [np.fmin(np.fmin(temperatura_tq, luz_le), i) for i in grande]
r2Uniao = [np.fmin(np.fmin(temperatura_tm, luz_lpe), i) for i in medio]
r3Uniao = [np.fmin(np.fmin(temperatura_tf, luz_ln), i) for i in baixo]
```

```python
print("União")
print('r1 :', r1Uniao)
print('r2 :', r2Uniao)
print('r3 :', r3Uniao)
```

```python
r1Intersecao = [np.fmax(np.fmin(temperatura_tq, luz_le), i) for i in grande]
r2Intersecao = [np.fmax(np.fmin(temperatura_tm, luz_lpe), i) for i in medio]
r3Intersecao = [np.fmax(np.fmin(temperatura_tf, luz_ln), i) for i in baixo]
```

```python
print("Interseção")
print('r1 :', r1Intersecao)
print('r2 :', r2Intersecao)
print('r3 :', r3Intersecao)
```

```python
# Como estamos utilizando a união e os valores de grande deram 0, serão utilizados os valores de baixo e medio
valoresPossiveis = [5, 20, 30, 35, 40, 60, 70]
valoresPossiveisIndices = [0.9, 0.6, 0.4, 0.5, 0.7, 0.7, 0.5]
```

```python
# Ativacao - Quantidade estimada de Turistas
qtd_lb = np.fmin(r1, LB) 
print(qtd_lb)
```

```python
qtd_lm = np.fmin(r2, LM) 
print(qtd_lm)
```

```python
qtd_la = np.fmin(r3, LA) 
print(qtd_la)
```

```python
#
qr1 = np.fmax(qtd_lb,qtd_lm)
ag_r1 = np.fmax(qr1,qtd_la)
print(ag_r1)
```
